```python
import re
import requests
from bs4 import BeautifulSoup as Soup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from urllib.request import urlopen
%matplotlib inline
```

```python
temporadas=['240025','230054','220069','210064','200061','190075','180084','170099']
```

```python
from urllib.request import urlopen, Request
filas_de_datos = []
for a in temporadas:
    url = "https://sofifa.com/teams?type=club&na%5B0%5D=45&lg%5B0%5D=53&r="+f"{a}"+"&set=true&showCol%5B%5D=oa&showCol%5B%5D=at&showCol%5B%5D=md&showCol%5B%5D=df&showCol%5B%5D=cw&showCol%5B%5D=ps"
    # Agregar un User-Agent en la solicitud
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    request = Request(url, headers=headers)

    # Realizar la solicitud con el User-Agent
    response = urlopen(request)

    # Leer el contenido de la página
    data = response.read()

    soup = Soup(data, 'html.parser')
    tabla = soup.find('table')
    encabezados = [header.text.strip() for header in tabla.find_all('th')]


    for fila in tabla.find_all('tr')[1:]: 
        datos_fila = [dato.text.strip() for dato in fila.find_all('td')]
        filas_de_datos.append(datos_fila)


df = pd.DataFrame(filas_de_datos, columns=encabezados)


```
